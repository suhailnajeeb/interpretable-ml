```python
# Path to dataset: 

data_path = 'Data/Bike-Sharing-Dataset/day.csv'
```

```python
# Library Imports
import pandas as pd
```

```python
#dateparse = lambda x: pd.datetime.strptime(x, '%Y-%m-%d %H:%M:%S')
dateparse = lambda x: pd.datetime.strptime(x, "%Y-%m-%d")
# Reference: https://stackoverflow.com/questions/17465045/can-pandas-automatically-recognize-dates
# df = pd.read_csv(infile, parse_dates=['datetime'], date_parser=dateparse)
```

```python
df = pd.read_csv(data_path, parse_dates = ['dteday'] , date_parser = dateparse)
```

```python
df.head()
```

```python
df.dtypes
```

```python
weekday_map = {0:'SUN', 1:'MON', 2:'TUE', 3:'WED', 4:'THU', 5:'FRI', 6:'SAT'}
holiday_map = { 0:'NO HOLIDAY', 1:'HOLIDAY'}
workingday_map = {0:'NO WORKING DAY', 1:'WORKING DAY'}
season_map = {0:'SPRING', 1:'SUMMER', 2:'FALL', 3:'WINTER'}
weathersit_map = {0:'GOOD', 1:'MISTY', 2:'RAIN/SNOW/STORM'}
mnth_map = {0:'JAN', 1:'FEB', 2:'MAR', 3:'APR', 4:'MAY', 5:'JUN', 6:'JUL', 7:'AUG', 8:'SEP', 9:'OKT', 10:'NOV', 11:'DEZ'}
yr_map = {0: '2011', 1: '2012'}
```

```python
df['weekday'] = df['weekday'].replace(weekday_map).astype("category")
df['holiday'] = df['holiday'].replace(holiday_map).astype("category")
df['workingday'] = df['workingday'].replace(workingday_map).astype("category")
df['season'] = df['season'].replace(season_map).astype("category")
df['weathersit'] = df['weathersit'].replace(weathersit_map).astype("category")
df['mnth'] = df['mnth'].replace(mnth_map).astype("category")
df['yr'] = df['yr'].replace(yr_map).astype("category")
```

```python
df.head()
```

```python
df.dtypes
```

```python
min_day = min(df['dteday'])
daydiff = lambda x : (x - min_day).days
```

```python
df['dteday'] = df['dteday'].apply(daydiff)
```

```python
df.head()
```

```python
df.dtypes
```

```python
# temp : Normalized temperature in Celsius. The values are derived via (t-t_min)/(t_max-t_min), t_min=-8, t_max=+39 (only in hourly scale)
df['temp'] = df['temp'] * (39 - (-8)) + (-8)
```

```python
# atemp: Normalized feeling temperature in Celsius. The values are derived via (t-t_min)/(t_max-t_min), t_min=-16, t_max=+50 (only in hourly scale)
df['atemp'] = df['atemp'] * (50 - (16)) + (16)
```

```python
#windspeed: Normalized wind speed. The values are divided to 67 (max)
df['windspeed'] = 67 * df['windspeed']
```

```python
#hum: Normalized humidity. The values are divided to 100 (max)
df['hum'] = 100 * df['hum']
```
